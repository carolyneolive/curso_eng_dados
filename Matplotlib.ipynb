```python
# Importando as funções da biblioteca
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Usa gráfico de linha
plt.figure(figsize=(5,2)) # 5 polegadas de largura e 2 de altura
plt.plot(np.array([4,6,10,9,6]))
plt.show()
```

```python
# Edição de eixo

# Alterando nomes dos eixos e acrescentando título
figura = plt.figure(figsize=(5,2))

# Colocando o título
plt.title("Função seno")

# Gerando valores para o seno
x = np.linspace(0, 3*np.pi, 100)  # Criando 100 pontos entre 0 e 3pi
seno = np.sin(x)  # Calculando o seno para cada ponto

# Plots de seno
plt.plot(x,seno, label='seno')
plt.plot(x,seno*-1, label='seno -1')

# Colocando os eixos
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

# Adicionando legenda
plt.legend(loc='lower right')

# Adicionando grade
plt.grid(True)
plt.show()
```

```python
# Alterando nomes dos eixos e acrescentando título
figura = plt.figure(figsize=(5,2))

# Gerando valores para o seno
x = np.linspace(0, 2*np.pi, 100)  # Criando 100 pontos entre 0 e 3pi
seno = np.sin(x)  # Calculando o seno para cada ponto

# Mudando cores e estilo de linhas e adicionando legenda
# Alterei 'x' para corresponder aos dados
plt.plot(x, seno, color='#C6A8C6', marker=',', label='seno')
# Alterei 'x' para corresponder aos dados, 'r--' para estilo de linha pontilhada vermelha
plt.plot(x, seno*-1, 'r--', label='1-seno')
plt.legend(frameon=False, loc='lower right')
# Adicionando grade
plt.grid(False)
plt.show()



# Marcadores:
# 'o': círculo
# 's': quadrado
# '^': triângulo para cima
# 'v': triângulo para baixo
# '>': triângulo para direita
# '<': triângulo para esquerda
# 'x': x
# '+': mais
# '*': estrela
# '.': ponto
# ',': pixel
```

## Gráfico de pizza

```python
# Usa gráfico de pizza
cidade = ["Porto de galinhas", "Jacumã", "Maragogi", "Pipa"]
populacao = [80.000, 24.323, 32.174, 13.609]
fig = plt.figure(figsize=(7,3))
plt.pie(populacao, labels=cidade, textprops=dict(fontsize=10, color="black"), autopct='%.1f%%')
#Leganda e títulos
titulo = "Praias do Nordeste"
plt.title(titulo)

plt.show()
```

```python
# Colocar a % e o número

def func(pct, allvals):
  absolute = int(np.round(pct/100.*np.sum(allvals)))
  return f"{pct:.1f}%\n Pop.:({absolute:d})"

cidade = ["Porto de galinhas", "Jacumã", "Maragogi", "Pipa"]
populacao = [80.000, 24.323, 32.174, 13.609]
fig = plt.figure(figsize=(7,3))
#Fatia da pizza que vou expandir
explode = (0.1,0,0,0)

plt.pie(populacao, labels=cidade, autopct=lambda pct:func(pct,populacao),textprops=dict(color="k", fontsize=9), explode=explode)


#Leganda e títulos
titulo = "Praias do Nordeste"
plt.title(titulo)

plt.show()
```

## Gráfico de Rosca

```python
explode = (0.05,0.05,0.05,0.05)

def func(pct, allvals):
  absolute = int(np.round(pct/100.*np.sum(allvals)))
  return f"{pct:.1f}%\n Pop.:({absolute:d})"

cidade = ["Porto de galinhas", "Jacumã", "Maragogi", "Pipa"]
populacao = [80.000, 24.323, 32.174, 13.609]
fig = plt.figure(figsize=(7,3))

plt.pie(populacao, labels=cidade, autopct=lambda pct:func(pct,populacao),textprops=dict(color="k", fontsize=9), explode=explode, pctdistance=0.75)

#Desenho de um círculo
circulo = plt.Circle((0,0), 0.60, fc="white")
fig= plt.gcf() #gcf acesso a figura atual, ou seja, ele pega a figura atual e faz e círculo em cima dela.

# Adicionando o círculo ao gráfico
# gca acesso a eixo atual, pega os eixos do gráfico de pizza.
fig.gca().add_artist(circulo) # add da nova figura

#Leganda e títulos
titulo = "Praias do Nordeste"
plt.title(titulo)

plt.show()
```

## Uso de datasets



```python
# Acesso ao googledrive dataset
df = pd.read_csv("/home/data.csv")
df.head()
```

```python
# Consulta dos tipos de variáveis do grupo.
df.info()
```

```python
# Pre-processamento dos dados
# Existe valores como NaN na coluna published_year
df['published_year'].fillna(0, inplace=True)

# Existe valores como NaN na coluna num_pages
df['num_pages'].fillna(0, inplace=True)

# Existe valores como NaN na coluna published_year
df['average_rating'].fillna(0, inplace=True)

# Existe valores como NaN na coluna num_pages
df['ratings_count'].fillna(0, inplace=True)
```

```python
# Converter a coluna para tipo int
df['published_year'] = df['published_year'].astype(int)
df['num_pages'] = df['num_pages'].astype(int)
df['average_rating'] = df['average_rating'].astype(int)
df['ratings_count'] = df['ratings_count'].astype(int)
```

```python
# Excluir as linhas com Zero na coluna 'published_year' e ano maior que 1975
df = df[(df['published_year'] != 0) & (df['published_year'] > 2007)]

# Excluir as linhas com Zero na coluna 'num_pages' e ano maior que 2000
df = df[(df['num_pages'] != 0) & (df['published_year'] > 2000)]

# Excluir as categorias vázias
df = df[pd.notna(df['categories'])]
```

```python
# Gráfico de barra com ano da publicação e a quantidade de páginas
plt.bar(df['published_year'],df['num_pages'], color='#E2A1F9') # barplot
# Gráfico de linha com ano da publicação e a quantidade de páginas
plt.plot(df['published_year'],df['num_pages'], color='k',linestyle='--') # lineplot

#Leganda e títulos
titulo = "O universo da leitura"
x_axis = "Ano de publicação"
y_axis = "Número de páginas"

plt.title(titulo)
plt.xlabel(x_axis)
plt.ylabel(y_axis)
```

```python
# Scatter
plt.scatter(df['published_year'],df['num_pages'])
plt.show()
```

```python
# Histograma
df.hist(figsize=(16,10),grid=False)
plt.show()
```
