###Pandas Vs Numpy

```python
import numpy as np
import pandas as pd

a = np.array([1,2,3,4])
series = pd.Series([1,2,3,4])

# Função não disponível no numpy
# Ela retorna o count,mean,std,min,max,quartil (25%,50%, 75%)
# Enquando em Numpy teria que fazer 1 por 1
print(series.describe(),'\n')

#Obs o acesso de elemento as duas biliotecas são iguais.
print('Numpy')
print(a[0])
print(a[:2],'\n')

print('Pandas')
print(series[0])
print(series[:2],'\n')
```

```python
# Funções bácicas
print('Numpy')
print('Média:        ',a.mean())
print('Desvio padrão:',a.std())
print('Máximo:       ',a.max())

print('Pandas')
print('Média:        ',series.mean())
print('Desvio padrão:',series.std())
print('Máximo:       ',series.max())
```

```python
# Funções bácicas

a = np.array([1,2,3,4])
b = np.array([1,2,1,2])

#Operação padrão
print('Numpy')
print('Soma:         ', a + b)
print('Subtração:    ', a - b)
print('Multiplicação:', a * b)
print('Divisão:      ', a / b)
print('Operação base:', a >= 3)
print('Operação base:', (a[a>= 3]))

a = pd.Series([1,2,3,4])
b = pd.Series([1,2,1,2])

#Operação padrão
# A diferença é que ele também trás o index
print('Pandas')
print('Soma:         ', a + b)
print('Subtração:    ', a - b)
print('Multiplicação:', a * b)
print('Divisão:      ', a / b)
print('Operação base:', a >= 3)
print('Operação base:', (a[a>= 3]))
```

```python
# No pandas eu consigo criar uma series e dar nome aos seus index o que não ocorre no numpy,
# pois para isso teria que criar dois array
# No Pandas temos a facilidade de agregação de informações

print('Numpy')
populacao = np. array([2.3333, 4.666554, 7.344554])
index = np.array(['Cabedelo', 'Bayeux', 'João Pessoa'])
print(populacao)
print(f'População de {index[1]} é de {populacao[1]} \n')

print('Pandas')
populacao =pd.Series([2.3333, 4.666554, 7.344554], index = ['Cabedelo', 'Bayeux', 'João Pessoa'])
print(populacao)
print(f'População de {index[1]} é de {populacao["Bayeux"]}')
```

```python
from sklearn.datasets import load_iris

# Carregar o conjunto de dados Iris
iris = load_iris()

#Fazemos isso, pois apos o load ele não vem como Dataframe essa linha é apenas para transformar da DT
features = pd.DataFrame(data=iris.data, columns=iris.feature_names)
features.head()

```

```python
# Utilização do Map
# Map só é aplicado em series.
# Em um dataframe você pode manipular apenas 1 coluna

# Função que transformad de centimetros para milimetros
def cm_to_mm(cm):
  mm = cm *10
  return mm

features['conversion_SL'] = features['sepal length (cm)'].map(cm_to_mm)
features.head()

# Como resultado do map temos a manipulação da coluna sepal length (cm).
```

```python
# Utilização do Apply
# Diferente do Map o Aplly pode ser aplicado em todo Dataframe
# Obs: Que seu quero selecionar apenas uma coluna eu coloco apenas 1 cochetes
#      Se quero selecionar multiplas colunas colocasse 2 cochetes
# Muito usado para conversão de tipo

features[['sepal length (cm)','sepal width (cm)']].apply(cm_to_mm).head()
```

```python
# Utilização do Applymap
# Ele vai aplicar a função no seu Dataframe interiro
features = features.drop('conversion_SL', axis=1) # Drop da coluna conversion_SL
features.applymap(cm_to_mm).head()
```

```python
#Soma
df1 = pd.DataFrame({'A':[1,2,3], 'B':[4,5,6], 'C':[7,8,9]})
df2 = pd.DataFrame({'A':[10,20,30], 'B':[40,50,60], 'C':[70,80,90]})
print(df1 + df2, '\n')

#Somando com overlap de colunas
df1 = pd.DataFrame({'A':[1,2,3], 'B':[4,5,6], 'C':[7,8,9]})
df2 = pd.DataFrame({'D':[10,20,30], 'C':[40,50,60], 'B':[70,80,90]})
print(df1 + df2, '\n')

#Somando com overlap de índice de linhas
df3 = pd.DataFrame({'A':[1,2,3], 'B':[4,5,6], 'C':[7,8,9]}, index=['row1','row2','row3'])
df4 = pd.DataFrame({'A':[10,20,30], 'B':[40,50,60], 'C':[70,80,90]},index=['row4','row3','row2'])
print(df3 + df4, '\n')
```

```python
#applymap() aplica a função a cada elemento do dataframe
df = pd.DataFrame({'A':[1,2,3], 'B':[10,20,30], 'C':[5,10,15]})

def add_one(x):
  return x + 1

print(df.applymap(add_one))

#Lambda

print(df.applymap(lambda x:x+1))
```

```python
#Aplicação de funções
df = pd.DataFrame({'A':[1,2,3], 'B':[10,20,30], 'C':[5,10,15]})

print(df.apply(np.mean))
print(df.apply(np.max))

def secound_larget(df):
  return df.nlargest(2).min() # nlargest é usado para obter as linhas com os menores/maiores valores em uma coluna específica de um DataFrame.
                              # Retorna os primeiros n Linhas ordenadas por colunas em ordem decrescente
print(df.apply(secound_larget))
```
